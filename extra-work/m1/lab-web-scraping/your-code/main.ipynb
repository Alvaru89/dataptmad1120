####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
html7=requests.get(url7)

print(html7.status_code)

soup = BeautifulSoup(html7.content, 'html.parser')

tbody = soup.find('tbody')
trs = tbody.find_all('tr')

res = []
for tr in trs:
    td = tr.find_all('td')
    row = [cell.text.strip() for cell in tr if cell.text]
    if row:
        res.append(row)

res
````

````python
#cleaning values
for row in res[:20]:
    if 'earthquake' not in row[0]:
        row.pop(0)
````

````python
colnames=['Time','Latitude','Orientation Lat', 'Longitude','Orientation Lon','Depth', 'Magnitude','Type','Region','Finishing time' ]
df=pd.DataFrame(res[:20], columns=colnames)
df
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url9 = 'https://www.wikipedia.org/'
````

````python
html9=requests.get(url9)

print(html9.status_code)

soup = BeautifulSoup(html9.content, 'html.parser')

cells = soup.find_all('a',{'class': 'link-box'})

import re

dict={}
for item in cells:
    lang=item.find('strong').text
    articles=item.find('bdi').text
    article=articles.replace('\xa0','.').replace('+', ' articles')
    dict[lang]=article

dict
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url11 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
html11=requests.get(url11)

print(html11.status_code)

soup = BeautifulSoup(html11.content, 'html.parser')

table = soup.find('table',{'class':'wikitable sortable'})

colnames=[]
ths = table.find_all('th')
for th in ths:
    colnames.append(th.text.strip())


res = []
trs = table.find_all('tr')
for tr in trs:
    td = tr.find_all('td')
    row = [cell.text.strip() for cell in td if cell.text]
    if row:
        res.append(row)

df=pd.DataFrame(res[:10], columns=colnames).set_index('Rank')
df
````
